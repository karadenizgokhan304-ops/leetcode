```python
from math import *
class Solution(object):
    def minTimeToVisitAllPoints(self, points):
        sec = 0
        for i in range(len(points) - 1):
            maxx = max(abs(points[i][0] - points[i+1][0]), abs(points[i][1] - points[i+1][1]))

            sec += maxx

        return sec #0 ms
        
```

```python
obj = Solution()

assert obj.minTimeToVisitAllPoints([[1,1],[3,4],[-1,0]]) == 7
```

```python
assert obj.minTimeToVisitAllPoints([[3,2],[-2,2]]) == 5
```
