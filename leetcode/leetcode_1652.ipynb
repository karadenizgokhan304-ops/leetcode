```python
class Solution:
    def decrypt(self, code: list, k: int) -> list:
        leng = len(code)
        total = 0
        l = []
        if k == 0:
            return [0]*leng
        
        if k>0:
            for i in range(leng):
                total = 0
                if i+k>=leng:
                    total += sum(code[:i+k-leng+1]) + sum(code[i+1:])
                else:
                    total+=sum(code[i+1:i+k+1])
                l.append(total)
        if k<0:
            for i in range(leng):
                total = 0
                if i-abs(k)<0:
                    total += sum(code[i-abs(k):]) + sum(code[:i])
                else:
                    total+=sum(code[i-abs(k):i])
                l.append(total)
        return l
```
