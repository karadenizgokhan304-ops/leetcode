```python
class Solution: #two set approach definitely bette
    def findRepeatedDnaSequences(self, s: str) -> list:
        hs = set()
        seq = set()
        for i in range(len(s)-9):
            seen = s[i:i+10]
            if seen in seq:
                hs.add(seen)
            seq.add(seen)
        return  list(hs)
```

```python
class Solution2: #one set approach
    def findRepeatedDnaSequences(self, s: str) -> list:
        hs = set()
        for i in range(len(s)-10):
            if s[i:i+10] in s[i+1:]:
                hs.add(s[i:i+10])
        return  list(hs)
```

```python
obj1 = Solution()  
obj2 = Solution()

assert obj1.findRepeatedDnaSequences("AAAAACCCCCAAAAACCCCCCAAAAAGGGTTT") == ["AAAAACCCCC","CCCCCAAAAA"]
assert obj1.findRepeatedDnaSequences("AAAAAAAAAAAAA") == ["AAAAAAAAAA"]
print("All tests are passed!!!")
```
