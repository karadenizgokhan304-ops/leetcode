```python
class Solution:
    def minOperations(self, nums: list) -> int:
        ans = 0
        for i in range(len(nums)-2):
            if nums[i] == 0:
                nums[i] = 1
                nums[i+1] = int(not nums[i+1]) #flipping, 1-nums alternative
                nums[i+2] = int(not nums[i+2])
                ans += 1
            else:
                continue
        if nums[-2] == 0 or nums[-1] == 0:  #at the end we must not see any "0" at the -1 or -2
            return -1
        return ans
        
```

```python
obj = Solution()
assert obj.minOperations([0,1,1,1,0,0]) == 3
assert obj.minOperations([0,1,1,1]) == -1
assert obj.minOperations([0,0,0]) == 1
assert obj.minOperations([1,0,0,1,1,1,0,1,1,1]) ==  3
print("All tests are passed!!!")


```
