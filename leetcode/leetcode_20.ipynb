```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        for i in range(0,len(s)):
            if s[i] in [")","]","}"] and not stack:
                return False
            if s[i] in ["(","[","{"]:
                stack.append(s[i])
            elif (stack[-1] == "(" and s[i] == ")") or (stack[-1] == "[" and s[i] == "]") or (stack[-1] == "{" and s[i] == "}"):
                stack.pop() 
            
            else:
                return False
        if not stack:
            return True  #return not stack alternative
        return False
```

```python
obj = Solution()

assert obj.isValid("()") == True
assert obj.isValid("()[]{}") == True
assert obj.isValid("(]") == False
assert obj.isValid("([])") == True
assert obj.isValid("([)]") == False

print("All tests are passed!!!!")
```
