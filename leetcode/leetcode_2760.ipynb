```python
from typing import List
class Solution:
    @classmethod
    def longestAlternatingSubarray(cl, nums: List[int], threshold: int) -> int:
        left = 0
        right = 0
        maxx = 0

        while right < len(nums):
            while left < len(nums) and (nums[left] % 2 != 0 or nums[left] > threshold):
                left += 1
                
            right = left
    
            while right < len(nums) and nums[right] <= threshold:
                if right > left: 
                    if nums[right] % 2 == nums[right - 1] % 2:
                        break
                maxx = max(maxx, right - left + 1)
                right += 1

            left += 1

        return maxx
```

```python
assert Solution.longestAlternatingSubarray(nums = [3,2,5,4],threshold = 5) == 3
assert Solution.longestAlternatingSubarray([1,2],2) == 1
assert Solution.longestAlternatingSubarray([1],1) == 0
assert Solution.longestAlternatingSubarray([2],2) == 1
assert Solution.longestAlternatingSubarray([1,6],2) == 0

print("All tests are passed!!!")
```
