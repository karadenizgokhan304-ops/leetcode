```python
from typing import List
from collections import Counter
class Solution:
    @classmethod #make it static func
    def longestNiceSubstring(self, s: str) -> str:
        if len(s) < 2:
            return ""
        lower = "abcdefghijklmnoprstuvyzqwx"
        upper = lower.upper()


        for char in s:
            if (char in lower and char.upper() not in s) or (char in upper and char.lower() not in s):
                return max((self.longestNiceSubstring(t) for t in s.split(char)),key = len)
        return s
```

```python
assert Solution.longestNiceSubstring("YazaAay") == "aAa"
assert Solution.longestNiceSubstring("Bb") == "Bb"
assert Solution.longestNiceSubstring("C") == ""


print("All tests are passed!!!")
```

```python
from collections import deque

class Solution2:
    @classmethod
    def longestNiceSubstring(self, s: str) -> str:
        if len(s) < 2:
           return ""
        queue = deque([s])
        longest = ""

        while queue:
            curr = queue.pop()
            char_set = set(curr)
            for i,c in enumerate(curr):
                if c.lower() not in char_set or c.upper() not in char_set:
                    for part in curr.split(c):
                        if len(part) > 0:
                            queue.append(part)
                    break
            
                elif len(curr) >= len(longest) and (i == len(curr)-1):
                    longest = curr
                    
        return longest
```

```python
assert Solution2.longestNiceSubstring("AaYzfgqBbYabA") == "Aa"
assert Solution2.longestNiceSubstring("Bb") == "Bb"
assert Solution2.longestNiceSubstring("C") == ""


print("All tests are passed!!!") 
```
