```python
class Solution:
    def simplifyPath(self, path: str) -> str:
        valid_path = []
        path = path.replace("//","/") #replacing "///" and "//" to "/ to create valid path
        path = path.replace("///","/")
        paths = path.split("/")  #splitting between "/" to creating a path list
        if paths[-1] == "":  #after we could have a "" empty string at the end of the list so we will pop it if exists
            paths.pop()
        paths.remove("") #remove first empty string at the beginning

        # print(path, paths)  for debugging
        for i in range(len(paths)):
            if paths[i] == "" or paths[i] == ".":
                continue
            if paths[i] == ".." and valid_path: #to avoid popping for empty list /../ exaple
                valid_path.pop()
                continue
            if not valid_path and paths[i] =="..": #just dont add any ".." into your path
                 continue
            else:
                valid_path.append(paths[i])
        return "/" + "/".join(valid_path) #list to string     summary: runtime 3ms beats %38 memory 17.5 mb beats %98
```

```python
obj = Solution()

assert obj.simplifyPath("/home/") == "/home"
assert obj.simplifyPath("/home//foo/") == "/home/foo"
assert obj.simplifyPath("/home/user/Documents/../Pictures") == "/home/user/Pictures"
assert obj.simplifyPath("/../") == "/"
assert obj.simplifyPath("/.../a/../b/c/../d/./") == "/.../b/d"

print("All tests are passed!!!!")
```
