```python
from math import gcd, lcm
class Solution:  #this question made me  psychologically unstable... the funny part of that,that was suggested as easy by creators... maybe ı should have improve myself
    def maxLength(self, nums: list) -> int:
        maxx = 0
        left = 0
        multiply = nums[0]
        curr_lcm = nums[0]
        curr_gbd = nums[0]
        if len(nums) == 2: #this was not necessary ı know
            return 2
        for right in range(1,len(nums)):
            curr_lcm = lcm(curr_lcm,nums[right])
            curr_gbd = gcd(curr_gbd,nums[right])
            multiply *= nums[right]
            while curr_lcm != multiply and curr_gbd*curr_lcm!=multiply:#multiply kinda useless ı guess gbd*lcm = multiply was enough
                multiply//=nums[left]
                curr_lcm = 1
                left+=1
                for i in range(left,right+1): #checking re-constructed window if it still has problems
                    curr_lcm = lcm(nums[i],curr_lcm)
            maxx = max(maxx,right-left+1) #final max lenght calculation
        return maxx #summary: runtime 7ms beats 90.31% memory usage 17.4mb beats 99.12%
                
                
```

```python
obj = Solution()
assert obj.maxLength([1,2,1,2,1,1,1]) == 5
assert obj.maxLength([2,3,4,5,6]) == 3
assert obj.maxLength([1,2,3,1,4,5,1]) == 5
assert obj.maxLength([6,7,4,2,2,1,1,1,4,5,5,5,6,7,8,1,1]) == 5 
print("All tests are passed!!!")


```
